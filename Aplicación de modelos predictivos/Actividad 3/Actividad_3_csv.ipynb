```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
import csv
```

```python
from collections import Counter
```

```python
plt.style.use("fivethirtyeight")

with open('data.csv') as csv_file:
  csv_reader = csv.DictReader(csv_file)
  languaje_counter = Counter()

  for row in csv_reader:
    languaje_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity = []

for item in languaje_counter.most_common(15):
  languages.append(item[0])
  popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity, color= '#493e63')

#plt.ylabel('Programming Languages')
plt.xlabel('Number of People Who Use', color= '#391140')
plt.title('Most Popular Languages', color='#493e63')



plt.tight_layout()


plt.show()
```
