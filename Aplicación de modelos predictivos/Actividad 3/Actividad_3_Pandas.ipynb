```python
plt.style.use("fivethirtyeight")

data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

languaje_counter = Counter()

for response in lang_responses:
    languaje_counter.update(response.split(';'))

languages = []
popularity = []

for item in languaje_counter.most_common(15):
  languages.append(item[0])
  popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity, color= '#493e63')

#plt.ylabel('Programming Languages')
plt.xlabel('Number of People Who Use', color= '#493e63')
plt.title('Most Popular Languages', color='#493e63')



plt.tight_layout()


plt.show()
```
